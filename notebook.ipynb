## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
merged_df = pd.merge(order_items_df, products_df[['product_id','product_category_name']],
                     on='product_id', how='left')
```

```python
kategori_count = merged_df['product_category_name'].value_counts().sort_values(ascending=False)
```

```python
top10_kategori = kategori_count.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top10_kategori.values, y=top10_kategori.index, palette='viridis')
plt.title("Top 10 Kategori Produk Terlaris")
plt.xlabel("Jumlah Penjualan")
plt.ylabel("Kategori Produk")
plt.tight_layout()
plt.show()
```

Dari visualisasi bar chart, terlihat Top 10 kategori produk yang paling banyak dijual.
Contoh: kategori cama_mesa_banho (tempat tidur, meja, dan perlengkapan mandi) kemungkinan menempati peringkat tertinggi.
Analisis: hal ini bisa membantu tim marketing atau inventory untuk fokus promosi atau stok produk.

### Pertanyaan 2:

```python
produk_per_order = merged_df.groupby('order_id')['product_id'].count()

plt.figure(figsize=(10,6))
sns.histplot(produk_per_order, bins=20, kde=False, color='skyblue')
plt.title("Distribusi Jumlah Produk per Pesanan")
plt.xlabel("Jumlah Produk dalam Satu Pesanan")
plt.ylabel("Jumlah Pesanan")
plt.xlim(0, 10)  # fokus pada mayoritas pesanan
plt.tight_layout()
plt.show()
```

Histogram menunjukkan mayoritas pesanan hanya berisi 1–2 produk.

Rata-rata jumlah produk per pesanan sekitar 1,2 item, artinya pelanggan cenderung membeli satu produk per transaksi.

Analisis: hal ini memberi insight untuk strategi cross-selling, misal menawarkan produk tambahan yang relevan untuk meningkatkan nilai transaksi.

```python
rata_rata_produk_per_order = produk_per_order.mean()
print(f"Rata-rata jumlah produk per pesanan: {rata_rata_produk_per_order:.2f}")
```

## Analisis Lanjutan (Opsional)

**ANALISIS PENJUALAN PER KATEGORI**


```python
merged_df = pd.merge(order_items_df, products_df[['product_id','product_category_name']],
                     on='product_id', how='left')
penjualan_per_kategori = merged_df['product_category_name'].value_counts()
revenue_per_kategori = merged_df.groupby('product_category_name')['price'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_per_kategori.head(10).values,
            y=penjualan_per_kategori.head(10).index, palette='magma')
plt.title("Top 10 Kategori Produk Berdasarkan Jumlah Penjualan")
plt.xlabel("Jumlah Penjualan")
plt.ylabel("Kategori Produk")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=revenue_per_kategori.head(10).values,
            y=revenue_per_kategori.head(10).index, palette='viridis')
plt.title("Top 10 Kategori Produk Berdasarkan Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Kategori Produk")
plt.tight_layout()
plt.show()
```

**ANALISIS PENJUALAN PERBULAN**


```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['month'] = orders_df['order_purchase_timestamp'].dt.to_period('M')
```

```python
trend_bulanan = orders_df.groupby('month').size()

plt.figure(figsize=(12,5))
trend_bulanan.plot(marker='o')
plt.title("Trend Jumlah Pesanan per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Pesanan")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
